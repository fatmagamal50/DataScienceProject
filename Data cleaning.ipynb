# outliers

Drop rows with zeros in the annual salary column

```python
print((df['annual salary'] == 0).sum())
df = df.drop(df[df['annual salary'] == 0].index)
```

drop all outliers

```python
# Calculate the summary statistics of the annual salary column
summary = df['annual salary'].describe()

# Calculate the interquartile range (IQR)
q1 = summary['25%']
q3 = summary['75%']
iqr = q3 - q1

# Calculate the upper and lower bounds for outliers
upper_bound = q3 + 1.5 * iqr
lower_bound = q1 - 1.5 * iqr
```

```python
# Identify outliers in the annual salary column
outliers = df[(df['annual salary'] < lower_bound) | (df['annual salary'] > upper_bound)]

# Drop the rows that contain outlier values
df = df.drop(outliers.index)
```
